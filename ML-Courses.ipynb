```python
import pandas as pd
```

```python
portuguese = pd.read_csv("student-por.csv", sep=',', low_memory=False)
portuguese
```

```python
math = pd.read_csv("student-mat.csv", sep=',', low_memory=False)
math
```

```python
course = pd.merge(left=math, right=portuguese, how='outer')
course
```

```python
course['sex'] = course['sex'].map({'F':0, 'M':1})
course['address'] = course['address'].map({'U':0,'R':1}) 
course['Pstatus'] = course['Pstatus'].map({'T':0,'A':1})
course['romantic'] = course['romantic'].map({'no':0,'yes':1})
course['internet'] = course['internet'].map({'no':0,'yes':1})
course['famsize'] = course['famsize'].map({'LE3':0,'GT3':1})
course['higher'] = course['higher'].map({'no':0,'yes':1})
course['school'] = course['school'].map({'GP':0,'MS':1})
```

```python
course = course.drop(['Medu','Fedu','Mjob','Fjob','studytime', 'traveltime', 'reason', 'nursery','freetime','paid','schoolsup', 'famsup','famrel','activities','G1','G2','G3','absences','goout','failures'], axis=1)
```

```python
attribute = 'guardian'
course = pd.concat([course,pd.get_dummies(course[attribute], prefix=attribute)],axis=1)
course.drop([attribute],axis=1, inplace=True)
```

```python
course
```

```python
courseX = course[['school','age','sex','address', 'Pstatus', 'higher','guardian_father','guardian_mother','guardian_other','internet']]
courseY = course['Dalc']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(courseX, courseY, test_size=0.2, shuffle=True, random_state=42, stratify=courseY)
```

```python
from sklearn.tree import DecisionTreeClassifier


decisionTree = DecisionTreeClassifier()
decisionTree.fit(X_train, y_train)

```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
# PrediÃ§Ã£o
predicao = decisionTree.predict(X_test)

# Matriz de confusÃ£o
print(confusion_matrix(y_test, predicao))

acuracia= accuracy_score(y_test, predicao)
print("AcurÃ¡cia: {:.2f}".format(acuracia))
```

```python
courseX = course[['address','Pstatus','age','Dalc', 'higher','guardian_father','guardian_mother','guardian_other']]
courseY = course['sex']
X_train, X_test, y_train, y_test = train_test_split(courseX, courseY, test_size=0.2, shuffle=True, random_state=42, stratify=courseY)
```

```python
from sklearn.neural_network import MLPClassifier
Rneural = MLPClassifier(alpha= 0.05, hidden_layer_sizes= (25, 35), learning_rate= 'constant', n_iter_no_change=5, power_t=0.8, random_state=42, shuffle=True, solver='adam')
Rneural.fit(X_train, y_train)
```

```python
parameter_space = {
    'hidden_layer_sizes': [(50,50), (25,75), (25,25)],
    'random_state':[1, 42],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'shuffle':[True, False],
    'n_iter_no_change':[5,10],
    'power_t':[0.2, 0.4, 0.6, 0.8],
    'learning_rate': ['constant','adaptive']
}

from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(Rneural, parameter_space, n_jobs=-1, cv=3)
clf.fit(X_train, y_train)

# Best paramete set
print('Best parameters found:\n', clf.best_params_)

# All results
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
```

```python
predicao = Rneural.predict(X_test)

# Matriz de confusÃ£o
print('Matriz de ConfusÃ£o')
print(confusion_matrix(y_test, predicao))

# AcurÃ¡cia
acuracia = accuracy_score(y_test, predicao)
print("AcurÃ¡cia: {:.2f}".format(acuracia))
```
